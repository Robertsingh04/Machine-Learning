# **One sample test**

```python
#Hypothesis one sample test- YearExperience

for i in range(7):
  sample = df.YearsExperience.sample(frac = 0.05)
  t_test, p_value = stats.ttest_1samp(sample, df.YearsExperience.mean())
  if p_value > 0.05:
    print("H0 Hypothesis null value is accepted, there is no significance difference")
  else:
    print("H0 Hypothesis null value is rejected, there is a significance difference")
```

```python
#Hypothesis one sample test- Salary

for i in range(7):
  sample = df.Salary.sample(frac = 0.04)
  t_test, p_value = stats.ttest_1samp(sample, df.Salary.mean())
  if p_value > 0.05:
    print("H0 Hypothesis null value is accepted, there is no significance difference")
  else:
    print("H0 Hypothesis null value is rejected, there is a significance difference")
```

# **Central Limit Theorem**

```python
#Central limit theorem: YearsExperiance

population = df.YearsExperience.values
res1 = population.mean()

samplesize = 5

sample_mean= []
for i in range(8):
  sample = np.random.choice(population, samplesize)
  sample_mean.append(sample)

res2 = np.mean(sample_mean)
print(res1,res2)
```

```python
#Central limit theorem: Salary

population = df.Salary.values
res1 = population.mean()

samplesize = 5

sample_mean= []
for i in range(8):
  sample = np.random.choice(population, samplesize)
  sample_mean.append(sample)

res2 = np.mean(sample_mean)
print(res1,res2)
```

# **Two sample Test**

```python
#Hypothesis two sample ttest - Price Vs Rating
for i in range(10):
  sample1 = df.YearsExperience.sample(frac=0.04)
  sample2 = df.Salary.sample(frac=0.05)
  t_value, p_value = stats.ttest_ind(sample1, sample2)

  if p_value > 0.05:
    print("H0 Null hypothesis is accepted and there is no significance difference between two columns")
  else:
    print("H0 Null hypothesis is rejected and there is a significance difference between two columns")

#H0 is accepted
```
