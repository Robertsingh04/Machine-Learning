```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.cluster import KMeans
```

```python
Mall_df=pd.read_csv("C:/Users/PC/Documents/DATA/Mall_Customers.csv")   
Mall_df.head()
```

```python
Mall_df["Gender"]=Mall_df["Gender"].replace({"Male":5,"Female":10})
```

```python
Mall_df
```

```python
sns.clustermap(Mall_df)
plt.savefig("tata.jpg")
```

```python
Mall_wcss=[]
for i in range(1,200):
  model=KMeans(i)
  model.fit(Mall_df)
  Mall_wcss.append(model.inertia_)



```

```python
Mall_wcss
```

```python
sns.set()
plt.plot(range(1,200),Mall_wcss)
plt.show()
```

```python
model=KMeans(10)
model.fit(Mall_df)
```

```python
identify_cluster=model.fit_predict(Mall_df)
identify_cluster
```

```python
Mall_df["cluster"]=identify_cluster
Mall_df
```

```python
Mall_df.cluster.value_counts()
```
